#Task 1: Find top 5 movies based on budget, revenue, profit, roi, vote average

##Tranform movieData set into another set that easy to perform this task

```python
movieDataTask1 = movieData[['title', 'poster_path', 'budget_musd', 'revenue_musd', 'vote_count', 'vote_average']].copy()
```

```python
movieDataTask1.set_index('title', inplace = True)
```

```python
movieDataTask1.columns = ['poster', 'budget', 'revenue', 'vote', 'average']
```

```python
movieDataTask1['profit'] = movieDataTask1['revenue'] - movieDataTask1['budget'] 
movieDataTask1['ROI'] = movieDataTask1['revenue']/movieDataTask1['budget'] 
```

```python
movieDataTask1.describe()
```

```python
movieDataTask1['budget'].fillna(0, inplace = True)
movieDataTask1['vote'].fillna(0, inplace = True)
```

```python
movieDataTask1.describe()
```

##Define a function for this task

```python
def bestWorseMovies(n, by, ascending = False, min_budget = 0, min_vote = 0):
  df = movieDataTask1.loc[(movieDataTask1['budget'] >= min_budget) & (movieDataTask1['vote'] >= min_vote), ['poster', by]].sort_values(by = by, ascending = ascending).head(n).copy()
  return HTML(df.to_html(escape = False))

```

##Find top 5 movies based on revenue





```python
bestWorseMovies(n = 5, by = 'revenue')
```

##Find top 5 movies based on budget


```python
bestWorseMovies(n = 5, by = 'budget')
```

##Find top 5 movies based on profit

```python
bestWorseMovies(n = 5, by = 'profit')
```

##Find top 5 movies based on ROI with minium budget is 50 M $

```python
bestWorseMovies(n = 5, by = 'ROI', min_budget = 50)
```

##Find top 5 movies based on average vote with minium vote is 50

```python
bestWorseMovies(n = 5, by = 'average', min_vote = 50)
```

#Task 2: Search movies based on certain conditions

##Search1: Science Fiction Action movie with Bruce willis (high rating)

```python
maskGenreActionFiction = movieData.genres.str.contains('Action') & movieData.genres.str.contains('Science Fiction')
```

```python
mastCastBruce = movieData.cast.str.contains('Bruce Willis')
```

```python
movieDataSearch1 = movieData.loc[maskGenreActionFiction & mastCastBruce, ['poster_path', 'title', 'vote_average']].sort_values(by = 'vote_average', ascending = False)
```

```python
HTML(movieDataSearch1.head().to_html(escape = False))
```

##Search2: Movie with uma thurman and directed by quentin tarantino (low runtime)

```python
maskCastUma = movieData.cast.str.contains('Uma Thurman')
```

```python
maskDirector = movieData.director == 'Quentin Tarantino'
```

```python
movieDataSearch2 = movieData.loc[maskCastUma & maskDirector, ['poster_path', 'title', 'runtime']].sort_values(by = 'runtime')
```

```python
HTML(movieDataSearch2.to_html(escape = False))
```

##Search 3: Most successful pixar studio movie between 2010 and 2015 (high revenue)

```python
maskStudio = movieData.production_companies.str.contains('Pixar').fillna(False)
```

```python
movieDataSearch3 = movieData['2010-01-01':'2015-12-31'].loc[maskStudio['2010-01-01':'2015-12-31'], ['poster_path', 'title', 'revenue_musd']].sort_values(by = 'revenue_musd', ascending = False)
```

```python
HTML(movieDataSearch3.head().to_html(escape = False))
```

##Search 4 Action or Thriller movie with original language English and min rating 7.5 (most recent)

```python
maskGenreActionThriller = movieData.genres.str.contains('Action') | movieData.genres.str.contains('Thriller')
maskLanguage = movieData.original_language == 'en'
maskVote = movieData.vote_count >= 10
maskRating = movieData.vote_average >= 7.5
```

```python
movieDataSearch4 = movieData.loc[maskGenreActionThriller & maskLanguage & maskVote & maskRating, ['poster_path', 'title', 'vote_average']].sort_index(ascending = False)
```
