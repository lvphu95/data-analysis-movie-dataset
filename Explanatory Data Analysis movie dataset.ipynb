#Task 1: Find top 5 movies based on budget, revenue, profit, roi, vote average

##Tranform movieData set into another set that easy to perform this task

```python
movieDataTask1 = movieData[['title', 'poster_path', 'budget_musd', 'revenue_musd', 'vote_count', 'vote_average']].copy()
```

```python
movieDataTask1.set_index('title', inplace = True)
```

```python
movieDataTask1.columns = ['poster', 'budget', 'revenue', 'vote', 'average']
```

```python
movieDataTask1['profit'] = movieDataTask1['revenue'] - movieDataTask1['budget'] 
movieDataTask1['ROI'] = movieDataTask1['revenue']/movieDataTask1['budget'] 
```

```python
movieDataTask1.describe()
```

```python
movieDataTask1['budget'].fillna(0, inplace = True)
movieDataTask1['vote'].fillna(0, inplace = True)
```

```python
movieDataTask1.describe()
```

##Define a function for this task

```python
def bestWorseMovies(n, by, ascending = False, min_budget = 0, min_vote = 0):
  df = movieDataTask1.loc[(movieDataTask1['budget'] >= min_budget) & (movieDataTask1['vote'] >= min_vote), ['poster', by]].sort_values(by = by, ascending = ascending).head(n).copy()
  return HTML(df.to_html(escape = False))

```

##Find top 5 movies based on revenue





```python
bestWorseMovies(n = 5, by = 'revenue')
```

##Find top 5 movies based on budget


```python
bestWorseMovies(n = 5, by = 'budget')
```

##Find top 5 movies based on profit

```python
bestWorseMovies(n = 5, by = 'profit')
```

##Find top 5 movies based on ROI with minium budget is 50 M $

```python
bestWorseMovies(n = 5, by = 'ROI', min_budget = 50)
```

##Find top 5 movies based on average vote with minium vote is 50

```python
bestWorseMovies(n = 5, by = 'average', min_vote = 50)
```

#Task 3 Find the most successful actor

```python
movieData.set_index('id', inplace =True)
```

```python
movieDataTask3 = movieData.cast.str.split('|', expand = True).stack().reset_index(level = 1, drop = True).to_frame()
```

```python
movieDataTask3.columns = ['Actor']
```

```python
movieDataTask3 = movieDataTask3.merge(movieData[['title', 'revenue_musd', 'vote_average', 'popularity']], how = 'left', left_index = True, right_index = True)
```

```python
movieDataTask3 = movieDataTask3.groupby('Actor').agg(Total_Revenue = ('revenue_musd', 'sum'),
                                                      Mean_Revenue = ('revenue_musd', 'mean'),  
                                                      Mean_Rating = ('vote_average', 'mean'),
                                                      Mean_Pop = ('popularity', 'mean'),
                                                      Total_Movies = ('Actor', 'count'))
                                                                                              
```

##Top 10 actors have the highest total movies acted

```python
totalMovie = movieDataTask3.Total_Movies.nlargest(10).to_frame()
```

```python
sns.barplot(data = totalMovie, x = totalMovie.index, y = 'Total_Movies')
plt.xticks(rotation=90)
plt.show()
```

##Top 10 actor have the highest total revenue

```python
totalRevenue = movieDataTask3.Total_Revenue.nlargest(10).to_frame()
```

```python
sns.barplot(data = totalRevenue, x = totalRevenue.index, y = 'Total_Revenue')
plt.xticks(rotation=90)
plt.show()
```

##Top 10 actors have the highest mean revenue




```python
meanRevenue = movieDataTask3[movieDataTask3.Total_Movies >= 10].Mean_Revenue.nlargest(10).to_frame()
```

```python
sns.barplot(data = meanRevenue, x = meanRevenue.index, y = 'Mean_Revenue')
plt.xticks(rotation=90)
plt.show()
```

##Top 10 actors have the highest mean vote average






```python
meanRating = movieDataTask3[movieDataTask3.Total_Movies >= 10].Mean_Rating.nlargest(10).to_frame()
```

```python
sns.barplot(data = meanRating, x = meanRating.index, y = 'Mean_Rating')
plt.xticks(rotation=90)
plt.show()
```

##Top 10 actors have the highest mean popularity


```python
meanPop = movieDataTask3[movieDataTask3.Total_Movies >= 10].Mean_Pop.nlargest(10).to_frame()
```

```python
sns.barplot(data = meanPop, x = meanPop.index, y = 'Mean_Pop')
plt.xticks(rotation=90)
plt.show()
```
